```python
## SRINIVAS RENGARAJAN
## R00183599

#Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#main program
def main():
    
    #Reading the dataset from csv file
    movieData=pd.read_csv("movie_metadata.csv")

    #Checking the dimensions of the data frame
    #print(movieData.shape)

    #Checking the structure of the dataset
    #movieData.info()

    #Checking the number of missing values in each column
    #print(movieData.isna().sum())

    ## Cleaning the data 
    ## Missing Value Rows Removal
    movieData=movieData.dropna()

    #Removing the special character (last character) from the movie titles
    movieData['movie_title']=movieData['movie_title'].map(lambda x: str(x)[:-1])
    #Checking the movie titles
    #print(movieData['movie_title'])
    
    #setting the plot background to white grid
    sns.set(style="whitegrid")
    userInput=1
    
    #User Interface for exploring IMDB dataset
    #The application continues to run till the user inputs 6th option
    while(userInput!=6):
        userInput=int(input("Please select one of the following options:\n1. Most successful actors or directors \n2. Film Comparison \n3. Analyse the distribution of gross earnings \n4. Genre Analysis \n5. Earnings and IMDB Scores \n6. Exit \n"))
        
        if userInput==1:
            selection=int(input("1. Top Directors \n2. Top Actors \n"))
            if selection==1:
                getTopDirectors(movieData)
            else:
                getTopActors(movieData)
        
        if userInput==2:
            isInValid=True
            
            while(isInValid):
                
                print("Enter two valid films from the dataset to compare the features: \n")
                film1=input("Film 1:")
                film2=input("Film 2:")
                isInValid=isFilmsInValid(film1,film2,movieData)
                
            if not isInValid:
                featureSelection=int(input("1. IMDB Scores \n2. Gross Earning \n3. Movie Facebook Like \n"))
                compareFilmsFeature(featureSelection,film1,film2,movieData)
                
        if userInput==3:
            
            isDataInvalid=True
            while(isDataInvalid):
                print("Enter two valid years:")
                #Get start year and end year to analyse gross earnings
                startYear=int(input("Enter Start Year:"))
                endYear=int(input("Enter End Year:"))
                isDataInvalid=isYearInValid(startYear,endYear,movieData)
                if not isDataInvalid:
                    
                    analyseGrossEarnings(movieData,startYear,endYear)
       
        if userInput==4:
            #Function to List all the unique genres 
            listUniqueGenres(movieData)
            #Get the user's choice of genre 
            genreChoice=input("Enter a specific Genre:")
            #Function to get the mean IMDB score of the films of the user selected genre 
            getGenreMeanIMDBScore(movieData,genreChoice)
            
        if userInput==5:
            #Function to plot the graphs to find/analyse the relation between imdb_score and other numerical columns 
            biVariateAnalysis(movieData)
                

def isFilmsInValid(film1,film2,movieData):
    #Case sensitivity is removed by converting both the user input and values in the column using lower() function
    if film1.lower() not in movieData['movie_title'].str.lower().unique() or film2.lower() not in movieData['movie_title'].str.lower().unique():
        return True
    else:
        return False

def isYearInValid(startYear,endYear,movieData):
    #Start year should be less than end year,otherwise ask the user to enter the year again
    if startYear < endYear:      
        #start year and end year must be available in the dataset
        if startYear not in movieData['title_year'].unique() or endYear not in movieData['title_year'].unique():            
            return True
        else:
            return False   
    else:
        return True

def compareFilmsFeature(featureSelection,film1,film2,movieData):
    #Case sensitivity is removed by converting both the user input and values in the column using lower() function 
    subset_films=movieData[(movieData.movie_title.str.lower()==film1.lower())|(movieData.movie_title.str.lower()==film2.lower())]
    
    #To plot bar graph for movie title and imdb scores 
    if featureSelection==1:
        sns.barplot(x="movie_title",y="imdb_score", data=subset_films)
        plt.show()
    
    #To plot bar graph for movie title and gross earnings 
    elif featureSelection==2:
        sns.barplot(x="movie_title",y="gross", data=subset_films)
        plt.show()
    
    #To plot bar graph for movie title and movie's facebook likes 
    else:
        sns.barplot(x="movie_title",y="movie_facebook_likes", data=subset_films)
        plt.show()
        
        
def getTopDirectors(movieData):
    numberOfDirectors=int(input("Number of Directors you want to return: \n"))
    if numberOfDirectors>0 and numberOfDirectors<=len(np.unique(movieData['director_name'])): 
        
        #Group all the directors and get the sum of their gross earnings
        sorted_movieData=movieData.groupby('director_name').agg({'gross':'sum'}).reset_index()
        
        #Sort the grouped data with descending order of gross earnings and get the inputted number of top directors
        sorted_movieData=sorted_movieData.sort_values('gross',ascending=False).head(numberOfDirectors)
        directors_plot=sns.barplot(x="gross", y="director_name", data=sorted_movieData)
       
        #directors_plot.set(ylabel="Gross in Billions")
        plt.show()
        
    else:
        print("Invalid/Wrong value!")
        getTopDirectors(movieData)
        
def getTopActors(movieData):
    numberOfActors=int(input("Number of Actors you want to return: \n"))
    if numberOfActors>0 and numberOfActors<=len(np.unique(movieData['actor_1_name'])):
        
        #Melting the data frame by creating one column called ActorName consisting three rows for every record or movie.
        actrGrp=pd.melt(movieData,id_vars=[i for i in movieData.columns if i not in ['actor_1_name','actor_2_name','actor_3_name'] ],var_name="Actor",value_name="ActorName")
        #print(actrGrp)
        
        
        #Group all the actors using ActorName column created and get the sum of their gross earnings
        sorted_movieData_actors=actrGrp.groupby(['ActorName']).agg({'gross':'sum'}).reset_index()
        #print(sorted_movieData_actors)
        
        #Sort the grouped data with descending order of gross earnings and get the inputted number of top actors
        sorted_movieData_actors=sorted_movieData_actors.sort_values('gross',ascending=False).head(numberOfActors)
        #print(sorted_movieData_actors)
        
        actors_plot=sns.barplot(x="gross", y="ActorName", data=sorted_movieData_actors)
        #actors_plot.set(ylabel='Gross in Billions')
        
        plt.show()
             
    else:
        print("Invalid/Wrong value!")
        getTopActors(movieData)
        
def analyseGrossEarnings(movieData,startYear,endYear):
    #Subset the moviedata dataframe in the start year and end year range
    yearSortedData=movieData[(movieData['title_year']>=startYear) & (movieData['title_year']<=endYear)]
    #print(yearSortedData.head())
    
    #Group by title_year column and find the aggregates such as minimum,mean and maximum for the gross earnings column
    year_wise_gross=yearSortedData.groupby('title_year').agg({'gross':['min','mean','max']})
    year_wise_gross.columns = year_wise_gross.columns.droplevel(level=0)
    year_wise_gross.rename(columns={
    "min": "min_gross", "max": "max_gross", "mean": "mean_gross"})
    #print(year_wise_gross)
    
    #Plot a line graph between the year and gross variables, the graph contains three lines, one for each min,mean and max gross for that year
    year_wise_gross.plot.line()
    #sns.lineplot(x='title_year',y='gross',data=year_wise_gross)
    plt.show()
    
def listUniqueGenres(movieData):
    print("List of unique Genres available:\n")
    
    #Split the genres with the '|' delimitter
    genreList=movieData['genres'].str.split("|")
    
    #Flatten the genrelist into a single list
    uniqueGenres = [genre for genres in genreList for genre in genres]
    
    #Display the unique genres alone
    print(np.unique(uniqueGenres))
    
def getGenreMeanIMDBScore(movieData,genreChoice):
    
    genreIndices=movieData['genres'].str.lower().str.contains(genreChoice.lower())
    #print(genreIndices)
    
    #Subset the indices from movieData which matches the user specified Genre
    genreData=movieData[genreIndices]
    #print(genreData.head())
    
    #Display the mean IMDB score of all the movies of the user specified Genre using pandas mean function
    print("The mean IMDB score of all the films in the specified genre is",genreData['imdb_score'].mean())
    
def biVariateAnalysis(movieData):
    #Select only the numeric columns (int and float types) alone from the movieData ''
    numericTypes=['int64','float64']
    numericMovieData = movieData.select_dtypes(include=numericTypes)
    #print(numericMovieData.columns)
    
    #Get the correlation matrix using corr() function
    correlation=numericMovieData.corr()
    
    print("Correlation between imdb_score and other numerical variables: \n")
    print(numericMovieData.corr()['imdb_score'])
    
    #Visualize the correlation using seaborn's heatmap
    sns.heatmap(correlation)
    plt.show()
    
    for i, column in enumerate(numericMovieData.columns):
        #Excluding imdb_score from x-variable to avoid plotting between imdb_score and imdb_score
        if column!='imdb_score':
            
            plt.figure(i)
            #Visualize the scatterplots to find the relation between imdb_score and other numeric variables
            sns.scatterplot(x=column, y='imdb_score', data=numericMovieData)
            plt.show()
    
main()    
```
